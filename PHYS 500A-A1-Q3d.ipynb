```python
%pylab inline
from scipy.optimize import *
```

```python
num=[0.2,0.1,0.01,0.001] # the constants of hbar^2/( m alpha L)
def eq(p):
    x, y = p
    return (((y*cos(2*x)-x*sin(2*x))*e**(-2*y)-y)/(x**2+y**2)-b\
            ,((x*cos(2*x)+y*sin(2*x))*e**(-2*y)-x)/(x**2+y**2) ) # the nonlinear function of real and imag parts
```

```python
m=[]
l=arange(0,11,1)    # the initial guesses of the value of k'
for i in range(len(l)):
    m.append(l[i])
```

```python
xx=np.zeros((len(num),len(m)))
yy=np.zeros((len(num),len(m)))
for i in range(len(num)):
    b=num[i]
    for j in range(len(m)):
        x, y =  fsolve(eq, (pi*m[j],0.01)) # find the numerical solutions and the initial guess of k" is 0 
        if x<=0:
            pass
        elif y>=0:
            pass
        else:
            xx[i,j]=x
            yy[i,j]=y
```

```python
for i in range(len(num)):
    figure(i)
    plt.plot(xx[i,1:],yy[i,1:],'ro')
    title(r'distribution of solutions in complex plane (m$\alpha$L/ $\hbar^2$ = %s)'%(1/num[i]))
    xlabel(' K\'L ')
    ylabel('k"L')
    xlim(0,32)
    show()
```
