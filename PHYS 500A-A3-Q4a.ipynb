```python
def f(NN):
    for i in range(len(NN)):
        N=NN[i]
        p=[]
        ap=[]
        x=linspace(0,N*3,N*3+1)
        y=linspace(0,N*3,1000)
        for j in range(len(x)):
            p.append((N/(1+N))**x[j]*(1/(1+N)))
        for k in range(len(y)):
            ap.append(1/(N)*e**(-y[k]/N))
        figure(i)
        plot(x,p,'bo')
        plot(y,ap,'r-',label='approximation')
        xlabel('photon number')
        ylabel('probability density')
        title(r'photon number distribution $<n>$=%s'%N)
        legend(loc='best')
```

```python
N=array([1,5,10,20,30,50])
f(N)
show()
```

### It can be seen that as the mean photon number grows, the approximation is getting better.
