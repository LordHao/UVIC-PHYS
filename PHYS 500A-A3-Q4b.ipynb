```python
%pylab inline
from scipy.special import factorial
```

```python
def P(m):
    p=[]
    G=[]
    
    sig=sqrt(m)
    if int(m-3*sig)>0:
        N=linspace(int(m-3*sig),int(m+3*sig),(int(m+3*sig)-int(m-3*sig)+1))
    else:
        N=linspace(0,int(m+3*sig),(int(m+3*sig)+1))
        
    NN=linspace(int(m-3*sig),int(m+3*sig),1000)
    for i in range(len(N)):
        n=N[i]
        p.append(e**(-m)*m**n/factorial(n))
    for i in range(len(NN)):
        n=NN[i]
        G.append(e**(-(n-m)**2/(2*m))/sqrt(2*pi*m))
    plot(N,p,'bo',label='P(n)')
    plot(NN,G,'r-',label='Gaussian approx')
    xlabel('photon number')
    ylabel('probability')
    title(r'the mean number of photon $\langle {n}\rangle$ = %s'%m)
    legend(loc='best')
```

```python
M=array([1,2,5,10,30])
```

```python
for i in range(len(M)):
    mm=M[i]
    P(mm)
    show()
```

### It can be seen that as the increase of the mean photon number $<n>$, the distribution of photon number follows a gaussian.
