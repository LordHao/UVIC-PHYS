```python
def Ec(z):
    for i in range(len(z)):
        Z=z[i]
        t=pi*linspace(0,8,100)
        dE=1/2
        y=real(Z)*cos(t)+imag(Z)*sin(t)
        yp=y+dE
        ym=y-dE
        figure(i)
        plot(t,y,'-') 
        plot(t,yp,'-')
        plot(t,ym,'-')
        xlabel(r'$\chi$')
        ylabel(r'amplitude scale by $\sqrt{2/\hbar}$')
        title('coherent state, Z=%s'%Z)
```

```python
def Es(Za,Phia):
    m=len(Za)
    n=len(Phia)
    fig, ax = subplots(m,n, figsize=(30,25))
    for i in range(m):
        for j in range(n):
            p=Phia[j]
            Z=Za[i]
            t=pi*linspace(0,4,1000)
            dE=sqrt((e**(-2*p)*cos(t)**2+e**(2*p)*sin(t)**2))/2
            y=e**(-1*p)*cos(t)*real(Z)+e**(p)*sin(t)*imag(Z)
            yp=y+dE
            ym=y-dE
            figure(i)
            ax[i,j].plot(t,y,'k-')
            ax[i,j].plot(t,yp,'-')
            ax[i,j].plot(t,ym,'-')
            ax[i,j].set_title(r'squeezed state, Z=%s, $\phi$=%s'%(Z,p),fontsize=20)
            ax[i,j].set_xlabel(r'$\chi$',fontsize=20)
            ax[i,j].set_ylabel(r'amplitude scale by $\sqrt{2/\hbar}$',fontsize=20)
    show()
            
```

```python
ZZ=array([0+0.1j,1,2+2j])
P=array([-1,1,-5,5])
```

```python
Ec(ZZ)
show()
Es(ZZ,P)
```
