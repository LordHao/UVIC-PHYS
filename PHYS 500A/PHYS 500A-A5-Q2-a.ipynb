```python
from scipy.special import spherical_jn as jl
from scipy.special import spherical_yn as nl
```

```python
L0=array([1,10,20,50,70,90,100])+1
```

```python
def sc(L):
    figure(figsize=(12,7))

    for i in range(len(L)):
        LL=L[i]
        x=linspace(0.1,100,1000)
        Tsc=[]
        for j in range(len(x)):
            xx=x[j]
            Tsc.append(sum([(2*l+1)*sin(arctan(jl(l,xx)/nl(l,xx)))**2 for l in range(LL)])/xx**2)
        plot(x,Tsc,label=r'truncate number $\scr{l}$=%s'%(LL-1))
        xlabel(r'$kR$',fontsize=15)
        ylabel(r'total cross section scaled by $\frac{1}{4 \pi R^2}$',fontsize=15)
        title(r'total cross section as a function of kR',fontsize=15)
        legend(loc='best')
    show()
```

```python
sc(L0)
```
