```python
from scipy.special import spherical_jn as jl
from scipy.special import spherical_yn as nl
from scipy.special import legendre 
```

```python
X0=array([0.1, 0.5, 1, 2, 10])
```

```python
def dl(n,xx):
    return arctan(jl(n,xx)/nl(n,xx))

def pl(n,x):
    pn=legendre(n)
    return pn(x)
```

```python
nn=51
def sc(X):
    for i in range(len(X)):
        x=X[i]
        theta=linspace(0,2*pi,400)
        Tsc=[]
        for j in range(len(theta)):
            t=theta[j]
            Tsc.append(abs(sum([(2*l+1)*sin(dl(l,x))*pl(l,cos(t))*\
                                exp(1j*dl(l,x)) for l in range(nn)]))/x**2)
        figure(i)
        figure(figsize=(9,5))
        if x==10:
            semilogy(theta,Tsc)
        else:
            plot(theta,Tsc)
        xlabel(r'$\theta$',fontsize=15)
        ylabel(r' differential cross section scaled by $\frac{1}{R^2}$',fontsize=15)
        title(r'differential cross section (kR=%s)'%x,fontsize=15)
    show()
```

```python
sc(X0)
```
