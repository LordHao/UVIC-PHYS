```python
Vf=[0.5,1,10,100]
```

```python
def TV(v):
    for i in range(len(v)):
        V=v[i]
        figure(figsize=(9,5))
        if v[i]==0:
            E=linspace(0.1,10,100)
        else:
            E=linspace(0.01,5,10000)*V
        T=4*E*(E+V)/(4*E*(E+V)+V**2*sin(2*(E+V)**0.5)**2)
        plot(E,T)
        xlabel(r'$\frac{2ma^2}{\hbar^2}E$',fontsize=15)
        ylabel('Transimission Probability',fontsize=15)
        title(r'Transimission of fixed $V_0$, ($\frac{2ma^2}{\hbar^2}V_0$= %s)'%V,fontsize=15)
    show()
```

```python
TV(Vf)
```

```python
Ef=0.0001
```

```python
figure(figsize=(9,5))
V=linspace(0,100,10000)
TE=4*Ef*(Ef+V)/(4*Ef*(Ef+V)+V**2*sin(2*(Ef+V)**0.5)**2)
plot(V,TE)
xlabel(r'$\frac{2ma^2}{\hbar^2}V_0$',fontsize=15)
ylabel('Transimission Probability',fontsize=15)
title(r'Transimission of $\frac{2ma^2}{\hbar^2}E$=%s'%Ef,fontsize=15)
show()
```
