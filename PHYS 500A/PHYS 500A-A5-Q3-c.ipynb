```python
def csV(v):
    for i in range(len(v)):
        V=v[i]
        figure(i)
        figure(figsize=(9,5))
        E=linspace(0.01,10,10000)*V
        cs= 2-(8*E*(E+V)*cos(2*sqrt(E+V))*cos(2*sqrt(E))+4*sqrt(E*(E+V))*\
               (2*E+V)*sin(2*sqrt(E+V))*sin(2*sqrt(E)))/(4*E*(E+V)+V**2*sin(2*sqrt(E+V))**2)
        plot(E,cs)
        xlabel(r'$\frac{2ma^2}{\hbar^2}E$',fontsize=15)
        ylabel('total cross section',fontsize=15)
        title(r'total cross section of $\frac{2ma^2}{\hbar^2}V_0$=%s'%V,fontsize=15)
    show()
```

```python
Vf=[0.5,1,10,100]
```

```python
csV(Vf)
```

```python
def csE(e):
    for i in range(len(e)):
        E=e[i]
        figure(i)
        figure(figsize=(9,5))
        V=linspace(0.01,100,10000)*E
        cs= 2-(8*E*(E+V)*cos(2*sqrt(E+V))*cos(2*sqrt(E))+4*sqrt(E*(E+V))*\
               (2*E+V)*sin(2*sqrt(E+V))*sin(2*sqrt(E)))/(4*E*(E+V)+V**2*sin(2*(E+V)**0.5)**2)
        plot(V,cs)
        xlabel(r'$\frac{2ma^2}{\hbar^2}V_0$',fontsize=15)
        ylabel('total cross section',fontsize=15)
        title(r'Total cross section of $\frac{2ma^2}{\hbar^2}E$=%s'%E,fontsize=15)
    show()
```

```python
Ef=[0.1,1,10,50]
```

```python
csE(Ef)
```
