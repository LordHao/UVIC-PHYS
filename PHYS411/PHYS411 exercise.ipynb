```python
#Q1
%pylab inline
from scipy.stats import chi2
```

```python
def xx1(tt1):
    return LL1*cos(ww_0*tt1) + AA1*cos(ww_1*tt1) + BB1*cos(ww_2*tt1)\
+ BC1*cos(ww_2c*tt1) + CC1* random.randn(NT)
```

```python
fsamp = 10
AA1 = 250
BB1 = 34
BC1 = 8
LL1 = 10
CC1 = 20
```

```python
#NT=20000
NT=20000
tt1 = linspace(1,NT,NT)/fsamp
TT1 = max(tt1)
fNq1 = NT/(2*TT1)
ww_1 = fNq1 *2*pi/100
ww_2 = 4.3 * ww_1
ww_2c = 4.7 * ww_1
ww_0 = fNq1 *2*pi/500
y=xx1(tt1)
plot(tt1,y)
xlabel('Time [s]')
ylabel('x [V]')
```

```python
XX1 = fft.fft(y)
XX1=array(XX1)
ii2 = linspace(1,NT/2,NT/2)
fq = (ii2-1)/TT1
fq=array(fq)
dfq = mean(diff(fq))
Gxx1=[]
for i in range(len(ii2)):
    Gxx1.append(abs((conj(XX1[i])*XX1[i])/(NT/2)))
Gxx1=array(Gxx1)
loglog(fq,Gxx1,linewidth=1)
xlabel('Frequency [Hz]')
ylabel('G_{xx} [V^2/Hz]')
```

```python
FS = 16
iiN = linspace(1,len(XX1),len(XX1))
figure(1)
plot(iiN,real(XX1),linewidth=1)
xlabel('index')
ylabel('Re(X)')
title('Discrete FFT')
figure(2)
plot(iiN,imag(XX1),linewidth=1)
xlabel('index')
ylabel('Im(X)')
```

```python
df = 2
cl_min = df / chi2.ppf(0.975, df)
cl_max = df / chi2.ppf(0.025,df)
fqL = 2
yyL = max(Gxx1)
loglog(fq,Gxx1,'r',linewidth=1.5)
loglog(dot(fqL,[1,1]),dot(yyL,[cl_min,cl_max]),'b',linewidth=2)
xlabel('Frequency [Hz]')
ylabel('G_{xx} [V^2/Hz]')
```

```python
Gxx1_min = df * Gxx1/chi2.ppf(0.975,df)
Gxx1_max = df * Gxx1/chi2.ppf(0.025,df)
loglog(fq,Gxx1,'r',linewidth=1.5)
loglog(dot(fqL,[1,1]),dot(yyL,[cl_min,cl_max]),'b',linewidth=2)
xlabel('Frequency [Hz]')
ylabel('G_{xx} [V^2/Hz]')
loglog(fq,Gxx1_min,'k--',fq,Gxx1_max,'k--',linewidth=0.5)
```

```python
#Q2
from scipy import signal
```

```python
#NT=20000
NT=20000
tt1 = linspace(1,NT,NT)/fsamp
TT1 = max(tt1)
fNq1 = NT/(2*TT1)
ww_1 = fNq1 *2*pi/100
ww_2 = 4.3 * ww_1
ww_2c = 4.7 * ww_1
ww_0 = fNq1 *2*pi/500
y=xx1(tt1)
NFFT1 = 2**14
NFFT2 = 2**9
pp = 0.95
LW1 = 1
LW2 = 0.5
```

fs: sampling rate;   nperseg: the length of each segment;   noverlap: If None, noverlap = nperseg/2. defaults to None;  nfft: Length of the FFT used, if a zero padded FFT is desired, if None, the FFT length is nperseg. defaults to None
;  default window is hanning.

```python
inter = chi2.interval(pp,df=2)

fq_M1 ,PP_M1= signal.welch(y,fs=fsamp, nperseg=NFFT1,noverlap=0,nfft=NFFT1)
figure(1)
subplot(211)
loglog(fq_M1,PP_M1,'b',linewidth=LW1)
fill_between(fq_M1,PP_M1/inter[0],PP_M1/inter[1],alpha=0.25,
                linewidth=0.0,edgecolor=None,color=None,facecolor='b')
LW1 = 1
fqLim = [10**(-4),10.0]
xlabel('Frequency [Hz]')
ylabel('G_{xx} [V^2/Hz]')
    
figure(2)
subplot(211)
fq_M2,PP_M2=signal.welch(y,fs=fsamp,nperseg=NFFT2)
loglog(fq_M2,PP_M2,'r',linewidth=LW1)
fill_between(fq_M2,PP_M2/inter[0],PP_M2/inter[1],alpha=0.25,
                linewidth=0.0,edgecolor=None,color=None,facecolor='b')
xlim(10**(-4),10.0)
xlabel('Frequency [Hz]')
ylabel('G_{xx} [V^2/Hz]')
```

```python
alpha_p = 1-pp
NS_1 = 2* floor(NT/NFFT1) - 1
nfree_1 = 2 * NS_1
cl_min_1 = nfree_1 / chi2.ppf(1-alpha_p/2,nfree_1)
cl_max_1 = nfree_1 / chi2.ppf(alpha_p/2,nfree_1)
NS_2 = 2* floor(NT/NFFT2) - 1
nfree_2 = 2 * NS_2
cl_min_2 = nfree_2 / chi2.ppf(0.975,nfree_2)
cl_max_2 = nfree_2 / chi2.ppf(0.025,nfree_2)
fqL_1 = 0.7
fqL_2 = 1.1
LW = 1
loglog(fq_M1,PP_M1,'b',fq_M2,PP_M2,'r',linewidth=LW)
loglog(dot(fqL_1,[1,1]),dot(yyL,[cl_min_1,cl_max_1]),'b',linewidth=2)
loglog(dot(fqL_2,[1,1]),dot(yyL,[cl_min_2,cl_max_2]),'r',linewidth=2)
xlabel('Frequency [Hz]')
ylabel('G_{xx} [V^2/Hz]')
```

the best choice for NFFT is NFFT1  ($2^{14}$), choose longer segements gives us more information, more details.
