$s^2=\displaystyle \frac{1}{N-1}\sum_{i=1}^{N}(x_i-\overline{x})^2$


$s_b^2=\displaystyle \frac{1}{N}\sum_{i=1}^{N}(x_i-\overline {x})^2$


$\displaystyle {\overline {x}}\,={\frac {1}{N}}\sum _{i=1}^{N}x_{i}\qquad$

$\displaystyle \operatorname {E}[\overline{x}]=\operatorname {E} \bigg[\frac{1}{N}\sum _{i=1}^{N}x_{i}\bigg]=\frac{1}{N}\operatorname {E} \bigg[\sum_{i=1}^{N}x_i\bigg]=\frac{1}{N}(N\mu)=\mu_{x}$ 


${\displaystyle {\begin{aligned}{\overline {x}}-\mu_{x} ={\frac {1}{N}}\sum _{i=1}^{N}x_{i}-\mu_{x} ={\frac {1}{N}}\sum _{i=1}^{N}x_{i}-{\frac {1}{N}}\sum _{i=1}^{N}\mu_{x} \ ={\frac {1}{N}}\sum _{i=1}^{N}(x_{i}-\mu_{x} )\\[8pt]\end{aligned}}}$


$\displaystyle \operatorname {E} \left[\frac{1}{N}\sum _{i=1}^{N}(x_i-\mu_{x} )^{2}\right]=\sigma_x^{2}$


$\displaystyle {\begin{aligned}\operatorname {E} \left[({\overline {x}}-\mu_{x} )^{2}\right]=\operatorname {E} \left[\bigg(\frac{1}{N}\sum _{i=1}^{N}(x_i-\mu_{x} )\bigg)^{2}\right]=\frac{1}{N}\operatorname {E} \left[\frac{1}{N}\sum _{i=1}^{N}(x_i-\mu_{x} )^{2}\right]=\frac{\sigma_x^{2}}{N}\end{aligned}} $

the expectation value of the cross product terms in the second expression is zero because the observations $x_i$ are independent 

${\displaystyle {\begin{aligned}\operatorname {E} [s_b^{2}]&=\operatorname {E} \left[{\frac {1}{N}}\sum _{i=1}^{N}{\big (}x_{i}-{\overline {x}}{\big )}^{2}\right]=\operatorname {E} {\bigg [}{\frac {1}{N}}\sum _{i=1}^{N}{\bigg (}(x_{i}-\mu_{x} )-({\overline {x}}-\mu_{x} ){\bigg )}^{2}{\bigg ]}\\[8pt]&=\operatorname {E} {\bigg [}{\frac {1}{N}}\sum _{i=1}^{N}{\bigg (}(x_{i}-\mu_{x} )^{2}-2({\overline {x}}-\mu_{x} )(x_{i}-\mu_{x} )+({\overline {x}}-\mu_{x} )^{2}{\bigg )}{\bigg ]}\\[8pt]&=\operatorname {E} {\bigg [}{\frac {1}{N}}\sum _{i=1}^{N}(x_{i}-\mu )^{2}-{\frac {2}{N}}({\overline {x}}-\mu_{x} )\sum _{i=1}^{N}(x_{i}-\mu_{x} )+({\overline {x}}-\mu_{x} )^{2}{\bigg ]}\\[8pt]&=\operatorname {E} {\bigg [}{\frac {1}{N}}\sum _{i=1}^{N}(x_{i}-\mu_{x} )^{2}-{\frac {2}{N}}({\overline {x}}-\mu_{x} )\cdot N\cdot ({\overline {x}}-\mu_{x} )+({\overline {x}}-\mu_{x} )^{2}{\bigg ]}\\[8pt]&=\operatorname {E} {\bigg [}{\frac {1}{N}}\sum _{i=1}^{N}(x_{i}-\mu_{x} )^{2}-2({\overline {x}}-\mu_{x} )^{2}+({\overline {x}}-\mu_{x} )^{2}{\bigg ]}\\[8pt]&=\operatorname {E} {\bigg [}{\frac {1}{N}}\sum _{i=1}^{N}(x_{i}-\mu_{x} )^{2}-({\overline {x}}-\mu_{x} )^{2}{\bigg ]}\\[8pt]&=\operatorname {E} {\bigg [}{\frac {1}{N}}\sum _{i=1}^{N}(x_{i}-\mu_{x} )^{2}{\bigg ]}-\operatorname {E} {\bigg [}({\overline {x}}-\mu_{x} )^{2}{\bigg ]}\\[8pt]&=\sigma_x^{2}-\frac{{\sigma_x}^{2}}{N}=\frac{N-1}{N}\sigma_x^{2}\end{aligned}}}$


$s_{b}^{2}$ is a biased estimator


$\displaystyle{ \begin{aligned}\operatorname {E} [s^{2}]=\operatorname {E} \bigg[\frac{N}{N-1}s_b^{2}\bigg]\end{aligned}}=\frac{N}{N-1}\operatorname {E} [s_b^{2}]=\sigma_x^{2}$

$s^{2}$ is an unbiased estimator

```python
%pylab inline
n=5            
for i in range (n):
    m=2**i           # m reprensents the number of the uniform PDFs are used to generate the new PDF
    figure(i)
    y=0
    for j in range(m): 
        x=rand(1000000)-0.5   # to generate the random numbers in the range of -0.5 to 0.5 for 1000000 times
        y=y+x
    y=y/m**0.5
    hist(y,bins=int(10*m**0.5),normed=1)
    title('probability density function %s'%(m)) 
    xlabel('events')
    ylabel('probability density ')
```

The more uniform distributions are added, the result looks more like a gaussian. The expectation value of Gaussian happens to be the average of the events which locates at the centre of the distribution, same as the uniformdistribution,  so we can simulate a gaussian by adding uniform distributions.

$p(x)=\frac{1}{11}\delta(x-2)+\frac{8}{11}\delta(x-3)+\frac{2}{11}\delta(x-4)$

probability distribution function

$\displaystyle \begin{aligned}P(x_0)&=\int_{-\infty}^{x_0} p(x)dx\\[8pt]&= \int_{-\infty}^{x_0}\bigg(\frac{1}{11}\delta(x-2)+\frac{8}{11}\delta(x-3)+\frac{2}{11}\delta(x-4)\bigg)dx \\[8pt]&=
\begin{cases}
0      & \quad x_0<2  \\
\frac{1}{11}  & \quad 2\leqslant x_0<3\\
\frac{9}{11}  & \quad 3\leqslant x_0<4\\
1  & \quad 4\leqslant x_0
  \end{cases}
\end{aligned}$

```python
def p(x):
    if any(x==2):
        return 1
    elif any(x==3):
        return 1
    elif any(x==4):
        return 1
    else:
        return 0
def P(x):
    if any(x<2):
        return 0
    elif any(2<=x<3):
        return N[0]/sum(N)
    elif any(3<=x<4):
        return (N[0]+N[1])/sum(N)
    elif any(4<=x):
        return 1
```

```python
m=5001
Y=[]
Z=[]
x=linspace(0,5,m)             
x=array(x)

for i in range(0,m):
    y=P(x[i])
    Y.append(y)
    z=p(x[i])
    Z.append(z)
    
Y=array(Y)
Z=array(Z)

figure(1)
xlim(1,5)
ylim(0,1)
plt.plot(x,Z,'red')
title('probability density function')
xlabel('events')
ylabel('probability density')
show()

figure(2)

plot(x,Y)
ylim(0,1.1)
title('probability distribution')
xlabel('events')
ylabel('probability')
show()
```
