```python
from scipy.special import factorial
```

$P_N(m)=\frac{N!}{((N+m)/2)!((N-m)/2)!}p^{(N+m)/2}(1-p)^{(N-m)/2}$

$\overline{m}=N(p-q)$

$\overline{(\Delta m)^2}=4Npq$

When  $n=\frac{1}{2}$

```python
def P(m):
    return factorial(N)/(factorial((N+m)/2.0)*factorial((N-m)/2.0))\
           *p**((N+m)/2.0)*(1-p)**((N-m)/2.0)
def G(x):
    return 1.0/sqrt(2*pi*s)*e**(-(x-mu)**2/(2.0*s)) # gaussian
```

```python
p=0.5
Ni=array([5,10,20])
```

```python
for j in range(len(Ni)):
    m=range(-Ni[j],Ni[j]+1,2)
    m=array(m)
    N=Ni[j]
    y=zeros(len(m))
    for i in range(len(m)):
        y[i]=P(m[i])
    mu=N*(2*p-1)
    s= 4*N*p*(1-p)
    x=linspace(min(m),max(m),1000)
    Gy=zeros(len(x))
    for k in range(len(x)):
        Gy[k]=G(x[k])
    figure(j+1)
    plot(m,y,'ro',label='Binomial')
    plot(x,Gy,'b',label='Gaussian')
    title(r'Binomial Distribution N=%s,p=%s,$\mu$=%s,$\sigma^2$=%s'\
    %(Ni[j],round(p,2),0,round(s,2)))
    legend(loc='best')
```

When  $n=\frac{2}{3}$

```python
p=2.0/3
```

```python
for j in range(len(Ni)):
    m=range(-Ni[j],Ni[j]+1,2)
    m=array(m)
    y=zeros(len(m))
    N=Ni[j]
    for i in range(len(m)):
        y[i]=P(m[i])
    mu=N*(2*p-1)
    s=4*N*p*(1-p)
    x=linspace(min(m),2*mu-min(m),1000)
    Gy=zeros(len(x))
    for k in range(len(x)):
        Gy[k]=G(x[k])
    figure(j+1)
    plot(m,y,'ro')
    title('Binomial Distribution N=%s,p=0.5'%Ni[j])
    figure(j+1)
    plot(m,y,'ro',label='Binomial')
    plot(x,Gy,'b',label='Gaussian')
    title(r'Probability Distribution N=%s,p=%s,$\mu$=%s,$\sigma^2$=%s'\
    %(Ni[j],round(p,2),round(mu,2),round(s,2)))    
    legend(loc='best')
```

The amplitube of Gaussian is only one half of the Binomial Distribution because $m$ assumes only integral values by $\Delta m=2$, the range $dx$ contains $dx/2$ possible values of $m$, all of which occur with the same probability $P_N(m)$.
