$p'=(p+\varepsilon \sin q)\%2\pi$

$q'=(p+q +\varepsilon \sin q)\%2\pi$

```python
def func(pt,ep): #define the Chirikov mapping
    p=pt[0]
    q=pt[1]
    pn=(p+ep*sin(q))%(2*pi)
    qn=(p+q+ep*sin(q))%(2*pi)
    return (pn,qn)
```

```python
def Ch(pt,ep):   #get the result
    arr=zeros((m+1,2))
    for i in range(m+1): #m is the number times of mapping
        arr[i]=pt
        pt=func(pt,ep)
    return arr
```

```python
g=linspace(0,2*pi,100)
pt0=array([(i,j) for i in g for j in g])
L=len(pt0)
shape(pt0)
```

```python
n=100
pp=linspace(0,2*pi,n)
qq=linspace(0,2*pi,n)
pt0= array([(i, j) for i in pp for j in qq])
print (L)
m=100  #number times of mapping
```

```python
import random
```

```python
N=4
sub=random.sample(range(L),N) #choose random N initial points
```

```python
pts=zeros((N,2))
for i in range(N):
    a=sub[i]
    pts[i]=pt0[a]
print(pts)
```

```python
eps=0.001
for i in range(N):
    figure(i)
    x=Ch(pts[i],eps)[:,0]
    y=Ch(pts[i],eps)[:,1]
    plot(x,y,'k.')
    xlabel('p')
    ylabel('q')
    title(r'initial point (%s,%s), $\varepsilon$=%s'%(round(pts[i,0],2),round(pts[i,1],2),eps))
```
