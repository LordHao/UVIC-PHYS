```python
%pylab inline
from scipy.integrate import odeint
```

(a)

```python
def func(r,u):  # define the second order differential equation
    x, vx, y, vy, z, vz = r
    drdu =  [vx, (1-vz)*sin(z-u),vy,0,vz,vx*sin(z-u)]
    return drdu
```

```python
r0 = [0,0,0,0,0,0] # initial state
u = linspace(0, 20*pi, 1000) 
traj = odeint(func, r0, u)
shape(traj)
```

```python
name=['x','vx','y','vy','z','vz'] 
tit =[r'trajectory in $\vec{E}$ direction ','',\
      r'trajectory in $\vec{B}$ direction ','',\
      r'trajectory in $\vec{k}$ direction ','']
```

```python

for i in range(len(traj[0,:])):
    if i%2==0:
        figure(i)
        figure(figsize=(7,7))
        plot(u,traj[:,i])
        xlabel(r'$ct $')
        ylabel('$%s $'%(name[i]))
        title(r'%s'%(tit[i]))
        show()
    else:
        pass
```

(b)

```python
g=1 # the damping constant
```

```python
def funcd(r, u):
    x, vx, y, vy, z, vz = r
    drdt =  [vx, (1-vz)*sin(z-u)-g*vx, vy,-vy,vz,vx*sin(z-u)-g*vz]
    return drdt
```

```python
r0 = [0,0,0,0,0,0]
u = linspace(0, 20*pi, 1000)
trajd = odeint(funcd, r0, u)
```

```python
for i in range(len(traj[0,:])):
    if i%2==0:
        figure(i)
        figure(figsize=(7,7))
        plot(u,trajd[:,i])
        xlabel(r'$ct $')
        ylabel('$%s $'%(name[i]))
        title(r'%s'%(tit[i]))
        show()
        
    else:
        pass
```
