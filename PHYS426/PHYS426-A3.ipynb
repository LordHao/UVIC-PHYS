```python
m=1
d=0.001

x, y = meshgrid(arange(-m, m+d, d),arange(-m, m+d, d))

u = y

v = x

figure(figsize=(8, 8))
streamplot(x, y, u, v, density=[0.4,0.5], color='k')
X=linspace(-1,1,1000)
Y=-X
plot(X,Y,'b-')
Y=X
plot(X,Y,'b-')
title(r'streamlines')
xlabel(r'$x$')
ylabel(r'$y$')
show()
```

```python
m=1
d=0.01
x=arange(-m, m+d, d)
y=arange(0, m+d, d)
xx, yy = meshgrid(x, y)
u = yy
v = xx
fu = interpolate.interp2d(x, y, u, kind='linear')
fv = interpolate.interp2d(x, y, v, kind='linear')
```

$\frac{\partial u}{\partial x}=\lim\limits_{\delta x \to 0}\frac{u(x+\delta x, y)-u(x,y)}{\delta x}$

$\frac{\partial v}{\partial y}=\lim\limits_{\delta y \to 0}\frac{v(x , y+\delta y) -v(x,y)}{\delta y}$

sum of linear strain rate : $\frac{\partial u_i}{\partial x_i}=0$

```python
a=23 #random x
b=34 # random y

pu=(u[b,a+1]-u[b,a])/d # parital du/dx, b represents row, which is y; a is x

pv=(v[b+1,a]-v[b,a])/d # parital dv/dy

print ('sum of the linear strain rate of raw data is:',(pu+pv) ) # sum of linear strain rate
```

```python
c=0.001
xnew = np.arange(-m, m+d, c)
ynew = np.arange(0, m+d, c )
U = fu(xnew, ynew) # interpolated u
V = fv(xnew, ynew) # interpolated v
```

```python
a=21 # random x 
b=36 # random y

pu=(U[b,a+1]-U[b,a])/c # parital du/dx, b represents row, which is y; a is x

pv=(V[b+1,a]-V[b,a])/c # parital dv/dy

print ('sum of the linear strain rate of interpolated data is:',(pu+pv) ) # sum of linear strain rate
```
