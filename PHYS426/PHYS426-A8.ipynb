```python
M=101
x=linspace(0,1,M)
```

```python
def fp(n,t,z):
    return 2/pi*e**(-n**2*pi**2*t)*(-1)**n/n*sin(n*pi*z)
```

```python
def ww(N,t,X):
    wp=zeros(M)
    u=zeros(M)
    nn=linspace(1,N,N)
    for i in range(M):
        z=X[i]
        for j in range(N):
            m=int(nn[j])
            wn=fp(m,t,z)
            wp[i]=wp[i]+wn
        u[i]=wp[i]+z
    return [wp,u]
```

```python
tt=array([0.01,0.05,0.2,0.5,1,10])
```

```python
for i in range(len(tt)):
    figure(i)
    figure(figsize=(8,8))
    T=tt[i]
    y=ww(1000,T,x)[0]
    u=ww(1000,T,x)[1]
    plot(x,y)
    xlabel('Z/H',fontsize=15)
    ylabel('w/U',fontsize=15)
    title(r'$\nu t/H^2=$ %s'%T,fontsize=20)
    xlim(0,1)
```

```python
for i in range(len(tt)):
    figure(i)
    figure(figsize=(8,8))
    T=tt[i]
    y=ww(1000,T,x)[0]
    u=ww(1000,T,x)[1]
    plot(x,y,label='w')
    plot(x,u,label='u')
    xlabel('Z/H',fontsize=15)
    ylabel('scaled by 1/U',fontsize=15)
    title(r'$\nu t/H^2=$ %s'%T,fontsize=20)
    xlim(0,1)
    legend(loc='best')
    show()
```

```python
L=101
```

```python
Tt=linspace(0,0.5,L)
zz=linspace(0,1,M)
ZZ,TT=meshgrid(zz,Tt)
Z=zeros((L,M))
Z1=zeros((L,M))
```

```python
for i in range(L):
    Z[i]=ww(1000,Tt[i],zz)[0]
    Z1[i]=ww(1000,Tt[i],zz)[1]
```

```python
figure(figsize=(10,7))
pcolormesh(ZZ, TT, Z, cmap='jet')
colorbar()
ylabel(r'$\nu t/ H^2$',fontsize=15)
xlabel(r'$Z/H$',fontsize=15)
title(r'$w/U$',fontsize=20)
show()
```

```python
figure(figsize=(10,7))
pcolormesh(ZZ, TT, Z1, cmap='jet')
colorbar()
ylabel(r'$\nu t/ H^2$',fontsize=15)
xlabel(r'$Z/H$',fontsize=15)
title(r'$u/U$',fontsize=20)
show()
```
