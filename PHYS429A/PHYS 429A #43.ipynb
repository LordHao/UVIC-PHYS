```python
def g(x,a,m,s):
    return a*exp(-(x-m)**2/(2*(s)**2))
def G(x, a, m, s,b,c,d):
    return a*exp(-(x-m)**2/(2*(s)**2))+b*x**2+c*x+d #modified gaussian distribution
def C(y,z):
    return sum((y-z)**2/z)
def L(x,m,b):
    return m*x+b
def S(x,a,b):
    return a+b*x**2
```

```python
y=array([139.55,237.42,302.60,354.34])
s=array([0.36,2.50,2.52,8.40])
print 'the peak channel numbers are:',y
print 'the uncertainty of the peak channel numbers are:',s
x=array([8.01,13.9,17.7,20.7])
para, unc = curve_fit(L,x,y,p0=[ 17.192, -1.599],maxfev=100000000)           
Z=array(L(x,*para))
plt.plot(x,y,'ro')
plt.plot(x,Z, 'b-', label='fit')
plt.ylabel('channel')
plt.xlabel('energy (kev)')
plt.title('Calibration')
chigs=(Z-y)**2/s**2
print 'the chi-square is:',round(sum(chigs),2)
m=round(para[0],2)
b=round(para[1],2)
ms=round(sqrt(diag(unc)[0]),2)
bs=round(sqrt(diag(unc)[1]),2)
plt.text(10,350, 'slope = %s +/- %s' % (round(para[0],2), round(sqrt(diag(unc)[0]),2)))
plt.text(10,330, 'intersect = %s +/-%s' % (round(para[1],2), round(sqrt(diag(unc)[1]),2)))
plt.text(10,310,'y= %s x + %s' % (round(para[0],2),round(para[1],2)))
plt.text(10,290,r'$\chi^2$ = %s'%(round(sum(chigs),2)))
plt.text(10,270,'d= %s'%(len(y)-2))
print 'the slope is:',para[0],'+/-',sqrt(diag(unc)[0])
print 'the intersect is:',para[1],'+/-',sqrt(diag(unc)[1])

```
