```python
def g(x,a,m,s,c):
    return a*exp(-(x-m)**2/(2*(s)**2))+c
def G(x, a, m, s,b,c,d):
    return a*exp(-(x-m)**2/(2*(s)**2))+b*x**2+c*x+d #modified gaussian distribution
def C(y,z):
    return sum((y-z)**2/z)
def L(x,m,b):
    return m*x+b
```

```python
M=array([0.1024,0.1007,0.1009,0.1011,0.1043])
Ms=array([0.0018,0.0029,0.0025,0.0021,0.0006])
B=array([8.76,8.43,7.88,6.62,7.06])
Bs=array([2.04,3.36,2.88,2.47,0.7])
Mm=round(mean(M),4)
Bm=round(mean(B),2)
MS=round((sum(Ms**2))**0.5/len(M),4)
BS=round((sum(Bs**2))**0.5/len(B),2)
print (Mm)
print (MS)
print (Bm)
print (BS)
```

```python
def Ch(x):
    return Mm*x+Bm
def E(x):
    return ((x-Bm)/Mm)/1000
def S(x,Ei,xs):
    return Ei*sqrt((xs**2+BS**2)/(x-Bm)**2+(MS/Mm)**2)
def St(x,Ei,xs):
    return Ei*sqrt((xs**2)/(x-Bm)**2)
```

```python
print (Ch(40000))
print (round((E(2400)),2))
print (round((E(5000)),2))
print (round((E(8191)),2))
```

```python
arr=genfromtxt("miniMuon.txt")
arr=arr.astype(int)
x_0 = []
y_0 = arr[:,1]
for i in range(len(arr)):
    x_0.append(E(arr[i,0]))

plot(x_0,y_0,'bo')
title('Muon Spectrum')
xlabel('energy(MeV)')
ylabel('counts')
xlim(E(20),80)
ylim(0,20)
show()
```

```python
nn=2400
mm=5000
ll=mm-nn+1
bs=100

x=linspace(nn,mm,ll)
y=arr[nn:(mm+1),1]
ii=linspace(nn+(mm-nn)/(1.0*bs),mm,bs)
y_y3=[]
x_x3=[]
for j in range(len(ii)):
    xx=[]
    yy=[]
    if j==0:
        for i in range(len(x)):
            if nn<x[i]<=ii[0]:
                yy.append(y[i])
                xx.append(x[i])
            else:
                pass
    else:
         for i in range(len(x)):
                if ii[j-1]<x[i]<=ii[j]:
                    yy.append(y[i])
                    xx.append(x[i])
                else:
                    pass
    if sum(yy)==0:  
        pass
    else:
        x_x3.append(dot(yy,xx)/sum(yy))
        y_y3.append(sum(yy))
y_y3=array(y_y3)
x_x3=array(x_x3)
```

```python
X=array(x_x3[:])
Y=array(y_y3[:])
a=amax(Y)
me= dot(X,Y)/sum(Y)
st=(sum(Y*(X-me)**2)/sum(Y))**0.5


para, unc = curve_fit(G, X, Y, p0=[a,me,st,0,-20,0],maxfev=1000000000)
z=array(G(X, *para))
yerr=sqrt(z)
errorbar(X,Y,yerr,fmt='o',color='blue')
plot(X, z, 'r-',label='fit')
title('Muon Energy Peak')
xlabel('Channel')
ylabel('counts')
legend(loc='best')
chiMu=C(Y,z)

print (round(para[0],2))
print (round(para[3],7))
print (round(para[4],2))
print (round(para[5],2))
Mu=round(para[1],2)
Mus=round(sqrt(diag(unc)[1]),2)

print ('the expectation value is:',Mu,'+/-',Mus)
print (r'std is:',round(para[2],2),'+/-',round(sqrt(diag(unc)[2]),2))
print ('the gaussian chisquare is :',round(chiMu,2))
print ('d is %s'%(len(X)-6))
```

```python
EMu=(round((E(Mu)),2))
Sta=round(St(Mu,EMu,Mus),2)
Sys=round(S(Mu,EMu,0),2)
SMu=round(S(Mu,EMu,Mus),2)
print ('the energy of muon peak is: %s MeV +/- %s'%(EMu,SMu))
print ('the energy of muon peak is: %s MeV +/- %s(stat) +/- %s (syst)'%(EMu,Sta,Sys))
```
