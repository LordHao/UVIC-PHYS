```python
%pylab inline
from scipy.optimize import curve_fit
```

```python
def L(x,mii,bii):
    return mii*x+bii
```

```python
arr=genfromtxt("cal44.txt")
arr=array(arr)
x=linspace(0,len(arr)-1,len(arr))
y=arr
plot(x,y,'b-')
xlabel('channel')
xlim(0,1024)
ylim(0,max(arr)*1.1)
ylabel('count')
title('calibration spectrum')
show()
```

```python
n=0
m=0
channel=[]
count=[]
num=[]
M=[]
sd=[]
for i in range (len(arr)):
    if arr[i]==0:
        pass
    else:
        channel.append(i)
    
ch=array(channel)
for j in range (len(ch)):
    if j==0:
        count.append(arr[channel[j]])
    elif j==len(ch)-1 :
        num.append(j+1)
        count.append(arr[channel[j]])
    else:
        if channel[j]>channel[j-1]+5:
            num.append(j)
            count.append(arr[channel[j]])
        else:
            count.append(arr[channel[j]])
    
co=array(count)
num=array(num)
for k in range(len(num)):
    if k==0:
        n=num[k]
        M.append(dot(co[0:n],ch[0:n])/sum(co[0:n]))
    else:
        m=num[k-1]
        n=num[k]
        M.append(dot(co[m:n],ch[m:n])/(sum(co[m:n])))
M=around(array(M),2)
print ('the mean values are: ',M)

for k in range(len(M)):
    if k==0:
        n=num[k]
        sd.append(((dot(co[0:n],(ch[0:n]-M[k])**2))/(sum(co[0:n])-1))**0.5)
    else:
        m=num[k-1]
        n=num[k]
        sd.append(((dot(co[m:n],(ch[m:n]-M[k])**2))/(sum(co[m:n])-1))**0.5)
sd=array(sd)
sd=(around(sd,2))
print ('the uncertainties are:',sd)
```

```python
st=0.64
NT=8.0
y=M
x=linspace(st,NT,len(num))
print (x)
plot(x,y,'ro')
xlabel('time ($\mu s$)')
ylabel('channel')
title(' Channel vs Time ')
para, unc = curve_fit(L,x,y,p0=[134.77, 0.01],maxfev=100000000)           
Z=array(L(x,*para))
plot(x,Z, 'b-')
chi=round(sum((Z-y)**2/(sd)**2),2)
m=round(para[0],2)
um=round(sqrt(diag(unc)[0]),3)
ins=round(para[1],4)
ui=round(sqrt(diag(unc)[1]),3)
text(0.5,800, 'slope = %s +/- %s' % (m,um))
text(0.5,700, 'intersect = %s +/-%s' % (ins, ui))
text(0.5,600,'y= %sx %s' % (m,ins))
text(0.5,500,r'$\chi^2$= %s' %chi)
text(0.5,400,'d=%s'%(len(x)-2))
xlim(0,8)
ylim(0,1024)
print (round(1/m,4))
```

```python
arr=genfromtxt("42.txt")
arr=array(arr)
def E(x,a,b,c):
    return a*exp(-(x/b))+c
def C(y,z):
    return (y-z)**2/y

def Ch(x):
    return m*x+ins
def T(x):
    return (x-ins)/m
def S(x,Ta,xs):
    return Ta*sqrt((xs**2+ui**2)/(x-ins)**2+(um/m)**2)
def St(x,Ta,xs):
    return Ta*sqrt((xs**2)/(x-ins)**2)
```

```python
x=linspace(0,len(arr)-1,len(arr))
x_1=1/m*(x-ins)
y=arr
print (sum(arr[178:]))
plot(x_1,y,'bo')
xlabel(r'$\mu s$')
ylabel('counts')
title('muon decay spectrum')
xlim(0,8)
ylim(0,max(arr[182:])*1.1)
pm=where(y==max(y[100:]))[0][0]
print(pm)
print(where(y==0)[0])
show()
```

```python
mm=978
nn=178
bs=80
ii=linspace(nn+(mm-nn)/(1.0*bs),mm,bs)
print ((mm-nn)/(1.0*bs))
print (ii)
y_y3=[]
x_x3=[]
for j in range(len(ii)):
    xx=[]
    yy=[]
    if j==0:
        for i in range(len(x)):
            if nn<x[i]<=ii[0]:
                yy.append(y[i])
                xx.append(x[i])
            else:
                pass
    else:
         for i in range(len(x)):
                if ii[j-1]<x[i]<=ii[j]:
                    yy.append(y[i])
                    xx.append(x[i])
                else:
                    pass
    if sum(yy)==0:
        pass
    else:
        x_x3.append(dot(yy,xx)/sum(yy))
        y_y3.append(sum(yy))
        
y_y3=array(y_y3)
x_x3=array(x_x3)
```

```python
x_3=array(x_x3[:])
y_3=array(y_y3[:])
para, unc = curve_fit(E,x_3,y_3,p0=[215.1,259,25],maxfev=100000000)  
a=round(para[0],2)
b=round(para[1],2)
c=round(para[2],2)
Z=array(E(x_3,*para))
yerr=array(sqrt(Z))
errorbar(x_3,y_3,yerr,fmt='o',color='b')
plot (x_3,Z,'r-',label='fit')
ua=round(sqrt(diag(unc)[0]),2)
ub=round(sqrt(diag(unc)[1]),2)
uc=round(sqrt(diag(unc)[2]),2)
chigs=around(C(y_3,Z),2)
print ('a is: %s +/- %s'%(a,ua))
print ('c is: %s +/- %s'%(c,uc))
print('b is:%s +/-%s'%(b,ub))
print ('chi-square is:',sum(chigs))
print ('df is:',len(y_3)-3)
text(500,175, r'y=%s $e^{-x/%s}$-%s' %(a,b,abs(c)))
text(500,150, r'$b$ = %s +/- %s'%(b,ub))
text(500,125,r'$\chi^2$= %s' %round(sum(chigs),2))
text(500,100,'d=%s'%(len(x_3)-3))
title('Muon decay curve (bins=%s)'%bs)
xlabel('channel')
xlim(nn-10,1000)
ylim(0,max(y_3+10))
ylabel('count')
legend(loc='best')
show()
```
