# Thomas BMT Equation

## $\frac{d\vec{s}}{dt} = \frac{q}{m \gamma} \vec{S} \times \big( (1+a\gamma)\vec{B}-a(\gamma - 1)(\hat{t}\cdot \vec{B})\hat{t}-(a\gamma + \frac{\gamma}{1+\gamma})\frac{\beta \hat{t}\times \vec{E}}{c}\big)$

### Where a is the anomalous g factor $a=\frac{g-2}{2};\hat{t}$ is the dirction aligned with the particle motion

### First, we are only aiming at the ring dipoles in the Rot region: B2E,BLA6RE. So the second and the third term vanish, also assume the vertical spin is 0.

## Now the equation reduced to:  $\frac{d\vec{s}}{dt}=\vec{\Omega} \times \vec{S} = (\frac{1}{\gamma}+a)\frac{q}{m} \vec{S} \times\vec{B_{\bot}} =-(\frac{1}{\gamma}+a)\frac{e }{m} B_{\bot} \vec{S}\times \hat{y}=(\frac{1}{\gamma}+a)\frac{e }{m} B_{\bot}(S_z \hat{x}-S_x \hat{z})$

##  $\vec{\Omega} = -\frac{q}{m \gamma}(1+a\gamma)\vec{B}_{\bot}$,  $\Omega$ is the axial vector which describes the spin precession(direction and the speed of precession)


## For electron, q = -e, where e is the elementary charge,  $\vec{\Omega} = \frac{e}{m \gamma}(1+a\gamma)\vec{B}_{\bot}$
## Note: For the negative charge particle like electron, the axial vector is always aligned with the external mangetic field (precession direction counterclockwise following the right hand rule) ; For positive charge particle, it's always anti-aligned (precession clockwise)



$ $
$ $

## Let $\omega=\frac{e}{m \gamma}(1+a\gamma)B_{\bot} $
## The equations of spin motion is given by :

## $\begin{cases} \begin{align*} \dot{S_x} &= \omega S_z \\  \dot{S_z}&=-\omega S_x\end{align*}\end{cases}$          


##  $\begin{cases} \begin{align*} \ddot{S_x} &= \omega \dot{S}_z = -\omega^2 S_x  \\  \ddot{S_z}&=-\omega \dot{S}_x = -\omega^2 S_z \end{align*}\end{cases}$

## This is the equation of harmonic oscillator, since we are only considering the fractional spin, the amplitude is 1, the solution is give by:

## $\begin{cases} \begin{align*} S_x &= \sin(\omega t + \phi)  \\  S_z&= \cos(\omega t+ \phi) \end{align*}\end{cases}$

## $\phi$ can be found by giving  the intial spin, $\phi = \arctan(\frac {S_{0x}}{S_{0z}})$ 

$  $


## When electron passes through the dipole field, the amount of spin rotation should be given by:   $\alpha=\omega\cdot \frac{L}{c}= (1+a\gamma)\frac{eL}{\gamma mc} B_{\bot}=(1+a\gamma)\frac{Lc}{\gamma E_0} B_{\bot} =(1+a\gamma)\frac{Lc}{E} B_{\bot} $

## when $\gamma =1$ (particle at rest), it reduces to Larmor precession

### Where L is the length of the dipole; $\alpha$ is the amount of rotation; E_0 is the rest mass energy of the electron, unit has to be eV

### It's easier to do the calculation if  both the numerator and the Denominator mutiply by c. In the Denominator, we will have m$c^2$ which has the unit of eV, so e in the numerator can be cancelled. The rest are all in SI unit.

## Now, consider the solenoid field, which is longitudinal 

## $\begin{align} \frac{d\vec{s}}{dt} &= \frac{q}{m \gamma} \vec{S} \times \big( (1+a\gamma)\vec{B}-a(\gamma - 1)(\hat{t}\cdot \vec{B})\hat{t} \big) \\ &= \frac{q}{m \gamma} \vec{S} \times \big( (1+a\gamma)(\vec{B}_{\bot}+\vec{B}_{//})-a(\gamma - 1)\vec{B}_{//} \big) \\ &= \frac{q}{m \gamma} \vec{S} \times \big( (1+a\gamma)\vec{B}_{\bot}+(1+a)\vec{B}_{//} \big)\\ &= -\frac{e}{m \gamma}\big((1+a\gamma) B_{\bot}\vec{S} \times \hat{y}+(1+a)B_{//}\vec{S} \times  \hat{z} \big)\\ & =-\frac{e}{m \gamma}\big((1+a\gamma) B_{\bot}(-S_z \hat{x}+S_x \hat{z})+(1+a)B_{//}(S_y \hat{x} - S_x \hat{y}) \big)\\ & = \frac{e}{m \gamma}\bigg[ \big((1+a\gamma)B_{\bot}S_{z}-(1+a)B_{//}S_{y} \big)\hat{x}+(1+a)B_{//}S_{x} \hat{y} - (1+a\gamma)B_{\bot}S_{x} \hat{z} \bigg] \end{align}  $

## we can conclude 3 differential equations of the spin motion:

## Let $\omega_B= \frac{e}{m \gamma}(1+a\gamma)B_{\bot}, \omega_S = \frac{e}{m \gamma}(1+a)B_{//} $

## $\begin{cases} \begin{align*} \dot{S_x} &= \omega_B S_z-\omega_{S}S_y \\ \dot{S_y} &= \omega_S S_x \\ \dot{S_z}&=-\omega_B S_x\end{align*}\end{cases}$

## $\begin{cases} \begin{align*} \ddot{S_x} &= \omega_B \dot{S}_z-\omega_{S}\dot{S}_y = -({\omega_B}^2+{\omega_S}^2) S_x \\ \ddot{S_y} &= \omega_S \dot{S}_x = -{\omega_S}^2 S_y + \omega_S \omega_B S_z \\ \ddot{S_z}&=-\omega_B \dot{S}_x= \omega_S \omega_B S_y-{\omega_B}^2 S_z \end{align*}\end{cases}$

## It's easy to solve the equation of motion for $S_x$

## Let $\omega_{SB}= \sqrt{{\omega_{S}}^2+{\omega_{B}}^2}$

## $S_x = A\sin\big(\omega_{SB}t + \phi \big)$


## y and z are coupled, use matrix to solve it

## $\big(\begin{smallmatrix}
  \ddot{S_y}\\
  \ddot{S_z}
\end{smallmatrix}\big)=\bigg(\begin{smallmatrix}
  -{\omega_{S}}^2 &\omega_{S}\omega_{B}\\
  \omega_{S}\omega_{B}& -{\omega_{B}}^2
\end{smallmatrix}\bigg)\big(\begin{smallmatrix}
  S_y\\
  S_z
\end{smallmatrix}\big) $

## Corresponding eigenvalues: $-{\omega_{S}}^2-{\omega_{B}}^2$ or 0

## The eigenvector of $-{\omega_{S}}^2-{\omega_{B}}^2$: $\frac{1}{\omega_{SB}}\big(\begin{smallmatrix}
  -\omega_{S}\\
  \omega_{B}
\end{smallmatrix}\big)$, for 0, it is: $\frac{1}{\omega_{SB}}\big(\begin{smallmatrix}
  \omega_{B}\\
  \omega_{S}
\end{smallmatrix}\big)$

## The general solution is:
## $S_y= -\frac{\omega_S}{\omega_{SB}} C\cos(\omega_{SB}t+\phi )+\frac{\omega_B}{\omega_{SB}}D$
## $S_z= \frac{\omega_B}{\omega_{SB}} C\cos(\omega_{SB}t+\phi )+\frac{\omega_S}{\omega_{SB}}D$



$ $
$ $


## First aim at a specific initial condition which is (0,1,0)

## At t = 0, the equations reduced to:
## $\begin{cases} \begin{align*} 0 &= A\sin\phi \\ 1 &=- \frac{\omega_S}{\omega_{SB}} C\cos\phi +\frac{\omega_B}{\omega_{SB}}D\\ 0&=\frac{\omega_B}{\omega_{SB}} C\cos\phi+\frac{\omega_S}{\omega_{SB}}D \end{align*}\end{cases}$

## First equation can give us $\phi = 0$, the equation of $S_y, S_z$ futher reduced to:

## $\begin{cases} \begin{align*} 1 &= -\frac{\omega_S}{\omega_{SB}} C +\frac{\omega_B}{\omega_{SB}}D\\ 0&=\frac{\omega_B}{\omega_{SB}} C +\frac{\omega_S}{\omega_{SB}}D \end{align*}\end{cases}$

## By solving the equations above, we can get $C=-\frac{\omega_S}{\omega_{SB}}, D=\frac{\omega_B}{\omega_{SB}}$

## One more constrain ${S_x}^2+{S_y}^2+{S_z}^2=1$ to find A, and $A=C=\frac{\omega_S}{\omega_{SB}}$ 

## Summarize the information above, the spin motion is described by:
## $\begin{cases} \begin{align*} S_x &= -\frac{\omega_S}{\omega_{SB}} \sin(\omega_{SB}t) \\ S_y&= \frac{{\omega_S}^2}{{\omega_{SB}}^2} \cos(\omega_{SB}t)+\frac{{\omega_B}^2}{{\omega_{SB}}^2}\\ S_z& = -\frac{\omega_S\omega_B}{{\omega_{SB}}^2} \big(1+\cos(\omega_{SB}t)\big) \end{align*}\end{cases}$


## Now consider the random initial condition($\tilde{S}_x,\tilde{S}_y,\tilde{S}_z$)
## At t = 0:
## $\begin{cases} \begin{align*} \tilde{S}_x &= A\sin\phi \\ \tilde{S}_y &= -\frac{\omega_S}{\omega_{SB}} C\cos\phi +\frac{\omega_B}{\omega_{SB}}D\\ \tilde{S}_z&=\frac{\omega_B}{\omega_{SB}} C\cos\phi+\frac{\omega_S}{\omega_{SB}}D \end{align*}\end{cases}$
## Use constrain  ${S_x}^2+{S_y}^2+{S_z}^2=1$ , can get A=C as before, but we can not solve for $\phi$ from the first equation

## Use the second and the third equation to eliminate $\phi$ term , can get $D=\frac{\omega_B\tilde{S}_y+\omega_S\tilde{S}_z}{\omega_{SB}}$

## Plug D into the 2nd or 3rd equation can get:   

## $C \cos\phi=\frac{\omega_{SB}\tilde{S}_x}{\omega_{B}\tilde{S}_z-\omega_S\tilde{S}_y}$

## Since we know $\tilde{S}_x = C\sin\phi$, then $\phi= \arctan\big(\frac{\omega_{SB}\tilde{S}_x}{\omega_{B}\tilde{S}_z-\omega_S\tilde{S}_y}\big)$


```python
E= 7007290000         # electron energy 7GeV
m= 510998.95          # rest mass energy of electron 511KeV
a= 0.001159652        # anomalous g factor for electron
c= 299792458          # speed of light
L= 5.902202           # length of both the B2E，BLA4lLE， BLA6RE, unit is metre
B= 0.22075143         # dipole strength of B2E in Tesla
ga= E/m               #lorentz boost fact
SLa = -4.843            #Solen strength B2EAL
SLb = -2.577           #Solen Srength B2EBL
SRa = -3.608           #Solen strength B2EAR
SRb = -3.942           #Solen Srength B2EBR
```

```python
def SR(B,S):
    Wb = (1+a*ga)*B*c/E
    Ws = (1+a)*S*c/E
    W = sqrt(Wb**2+Ws**2)
    return Wb,Ws,W
```

```python
def S_G(B,S,si,s):       #to find spin at the exit of the dipole and the solenoid combined function magnet
    
    six = si[0]        # intial spin.x
    siy = si[1]        # intial spin.y
    siz = si[2]        # intial spin.z
    
    wb,ws,w = SR(B,S)
         
    D = (wb*siy+ws*siz)/w  # find D
    
    phi = math.atan2(w*six,(-ws*siy+wb*siz))        # find phi
    
    #print('the amount of rotation is: %s degrees'%(w*s*180/pi))
            
    if abs(phi)==pi/2:          #if cos(phi)=0, the calculation of C below would not work
        C=six
    else:
        C=(-ws*siy+wb*siz)/(w*cos(phi))   
        
    sx =  C*sin(w*s+phi)
    
    sy = -ws/w*C*cos(w*s+phi)+wb/w*D
    
    sz =  wb/w*C*cos(w*s+phi)+ws/w*D
    
    return sx,sy,sz
```

```python
s0= array([0, 1, 0])        #initial spin which is in the y direction
```

```python
sg=S_G(B,SLb,s0,L)  # when particle passing through the B2EBL spin rotator 
print('the spin at the exit of B2EBL rot',sg)
```

```python
sg=S_G(B,SLa,sg,L)  # when particle passing through the B2EAL spin rotator 
print('the spin at the exit of second B2EAL rot',sg)
```

```python
s0= array([0, 1, 0]) 
wblb,wslb,wlb = SR(B,SLb)
slb = arange(0, 2*pi/wlb+0.1, 0.1)
Sglb=S_G(B,SLb,s0,slb)  # when particle passing through the B2EBL spin rotator

slbL = arange(0, L+0.1, 0.1)
SglbL=S_G(B,SLb,s0,slbL)  # when particle passing through the B2EBL spin rotator 
```

```python
s0 = array([SglbL[0][-1],SglbL[1][-1],SglbL[2][-1]])
wbla,wsla,wla = SR(B,SLa)
sla = arange(0, 2*pi/wla+0.1, 0.1)
Sgla=S_G(B,SLa, s0,sla)  # when particle passing through the B2EAL spin rotator 

slaL = arange(0, L+0.1, 0.1)
SglaL=S_G(B,SLa,s0,slaL)  # when particle passing through the B2EAL spin rotator 
```

```python
fig = figure(figsize=(13,11))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(Sglb[2],Sglb[0],Sglb[1], color= 'olive', marker='.',s=30)
ax.scatter(Sgla[2],Sgla[0],Sgla[1], color= 'olive', marker='.',s=30)
ax.plot(SglbL[2],SglbL[0],SglbL[1],'r-',linewidth = 5 )
ax.plot(SglaL[2],SglaL[0],SglaL[1],'r-',linewidth = 5 )
ax.quiver(zeros(len(SglbL[2][1::5])), zeros(len(SglbL[2][1::5])), zeros(len(SglbL[2][1::5])), \
          SglbL[2][1::5],SglbL[0][1::5],SglbL[1][1::5],color ='darkorange',alpha = 0.5)

ax.quiver(zeros(len(SglaL[2][::10])), zeros(len(SglaL[2][::10])), zeros(len(SglaL[2][::10])), \
          SglaL[2][::10],SglaL[0][::10],SglaL[1][::10],color ='darkorange',alpha = 0.5)

ax.quiver(0, 0, 0, SglaL[2][-1],SglaL[0][-1],SglaL[1][-1],color ='green',alpha = 1)

ax.quiver(0, 0, 0, 0, 0, 1, color ='blue',alpha = 1)
ax.quiver(0, 0, 0, wsla/wla, 0 , wbla/wla,color ='k',alpha = 1)
ax.quiver(0, 0, 0, wslb/wlb, 0 , wblb/wlb,color ='k',alpha = 1)


    
ax.set_xlim([-1, 0])
ax.set_ylim([-0.6,0.6])
ax.set_zlim([0, 1])
    
#ax.set_title(r'Spin Motion of $e ^-$ (Lab Frame) in the SuperKEKB HER with Spin Rotator Installed', size =17) 
ax.set_xlabel(r'Z ',fontsize=15)
ax.set_ylabel(r'x ',fontsize=15)
ax.set_zlabel(r'Y',fontsize=15)
show()
```
